```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("salary.csv")
x=np.array(df[['x']])
y=np.array(df[['y']])
x,y
```

```python
plt.scatter(x,y)
plt.xlabel("Cgpa")
plt.ylabel("Package")
plt.show()
```

```python
x_mean=np.mean(x)
y_mean=np.mean(y)
print(x_mean,y_mean)
```

```python
num=0
den=0
for i in range(len(x)):
    num+=(x[i]-x_mean)*(y[i]-y_mean)
    den+=(x[i]-x_mean)**2
    b1=num/den
    b0=y_mean-(b1*x_mean)
print("intercept",b0)
print("cofficient",b1)
```

```python
y_pred=b0+b1*x
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred)
plt.show()
```

```python
num_r=0
den_r=0
for i in range(len(x)):
    num_r+=(y[i]-y_pred[i])**2
    den_r+=(y[i]-y_mean)**2
    r_square=1-num_r/den_r
print("value of coefficient of determination R_sqaure",r_square)
```
